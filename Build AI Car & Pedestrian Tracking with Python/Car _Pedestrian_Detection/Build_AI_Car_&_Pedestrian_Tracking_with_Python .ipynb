# Detecting with cascades 

```python
car_cascade = cv2.CascadeClassifier('cars.xml')
img = cv2.imread('highway-5c6e5194c9e77c0001cda269.jpg')
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY) # to convert the image to white to black

# detect the cars 
cars = car_cascade.detectMultiScale(gray , 1.1 , 1)

# Draw the Border
for(x,y,w,h) in cars:
    cv2.rectangle(img,(x,y) , (x+w , y+h) , (0,0,255) , 2)

# resizing the image
resized = cv2.resize(img , (int(img.shape[1]/2) , int(img.shape[0]/2)))

cv2.imshow("car",resized)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

# Detecting the Pedestrains 

```python
import cv2,time
pedestrian_detector = cv2.CascadeClassifier('pedestrian.xml')
video = cv2.VideoCapture('Pedestrians Compilation 480 x 854.mp4')

a = 1
while True:
    a = a+1
    check,frame = video.read()
    print(frame)
    gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    pedestrians = pedestrian_detector.detectMultiScale(gray , 3.5 , 1)
    cars = car_cascade.detectMultiScale(gray , 1.1 , 1)
        
    for (x,y,w,h) in pedestrians:
        cv2.rectangle(frame,(x,y) , (x+w , y+h) , (255,0,0) , 2)
    
    for (x,y,w,h) in cars:
        cv2.rectangle(frame,(x,y) , (x+w , y+h) , (0,0,255) , 2)
    resize = cv2.resize(frame, (640, 480)) 
    cv2.imshow("car", resize) 
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
print(a)
video.release()
cv2.destroyAllWindows()
```

# Detecting Both Pedestrian And The Cars in the video
